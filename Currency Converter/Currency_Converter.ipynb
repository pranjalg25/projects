```python
import tkinter as tk
from tkinter import messagebox
from forex_python.converter import CurrencyRates

def real_time_currency_conversion():
    from_currency = variable1.get()
    to_currency = variable2.get()

    if amount1_field.get() == "":
        messagebox.showinfo("Error !!", "Amount Not Entered.\n Please enter a valid amount.")
    elif from_currency == "currency" or to_currency == "currency":
        messagebox.showinfo("Error !!", "Currency Not Selected.\n Please select FROM and TO Currency from the menu.")
    else:
        c = CurrencyRates()
        new_amt = c.convert(from_currency, to_currency, float(amount1_field.get()))
        new_amount = float("{:.4f}".format(new_amt))
        amount2_field.delete(0, tk.END)
        amount2_field.insert(0, str(new_amount))

def clear_all():
    amount1_field.delete(0, tk.END)
    amount2_field.delete(0, tk.END)

# GUI
root = tk.Tk()
root.title("Currency Converter")
root.configure(background='#262526')  # Set background color

# Top Frame
tops = tk.Frame(root, bg='#262526', pady=10, width=1850, height=100, relief="ridge")  # Add padding
tops.grid(row=0, column=0)

headlabel = tk.Label(tops, font=('lato black', 19, 'bold'), text='Currency Converter', bg='#262526', fg='white')  # Change background color
headlabel.grid(row=1, column=0, sticky=tk.W, padx=10)  # Add padding

# Set default currency
variable1 = tk.StringVar(root)
variable2 = tk.StringVar(root)
variable1.set("currency")
variable2.set("currency")

# Currency codes
currency_code_list = ["INR", "USD", "CAD", "CNY", "DKK", "EUR"]

# Main Frame
root.geometry("700x400")


label_1 = tk.Label(root, font=('lato black', 27, 'bold'), text="", padx=2, pady=2, bg="#262526", fg="white")
label_1.grid(row=1, column=0, sticky=tk.W)

amount1_label = tk.Label(root, font=('lato black', 15, 'bold'), text="Enter Amount : ", bg="#262526", fg="orange")
amount1_label.grid(row=1, column=0, sticky=tk.W, pady=2)  # Adjust row position
amount1_field = tk.Entry(root)
amount1_field.grid(row=2, column=0, ipadx=28, sticky=tk.E, pady=2)  # Adjust row position

labels = ["From Currency", "To Currency", "Converted Amount"]
for i, text in enumerate(labels, start=2):
    label = tk.Label(root, font=('lato black', 15, 'bold'), text=f"{text} : ", bg="#262526", fg="orange")  # Change text color
    label.grid(row=i * 2 - 1, column=0, sticky=tk.W, pady=2)  # Adjust row position

from_currency_option = tk.OptionMenu(root, variable1, *currency_code_list)
to_currency_option = tk.OptionMenu(root, variable2, *currency_code_list)
from_currency_option.grid(row=4, column=0, ipadx=45, sticky=tk.E, pady=2)  # Adjust row position
to_currency_option.grid(row=6, column=0, ipadx=45, sticky=tk.E, pady=2)  # Adjust row position


amount2_label = tk.Label(root, font=('lato black', 15, 'bold'), text="Converted Amount : ", bg="#262526", fg="orange")
amount2_label.grid(row=10, column=0, sticky=tk.W, pady=2)  # Adjust row position
amount2_field = tk.Entry(root)
amount2_field.grid(row=10, column=0, ipadx=31, sticky=tk.E, pady=2) 

convert_button = tk.Button(root, font=('arial', 15, 'bold'), text=" Convert ", padx=2, pady=2, bg="#262526", fg="white", command=real_time_currency_conversion)  # Change background color
convert_button.grid(row=15, column=0, pady=10)  # Add padding

clear_button = tk.Button(root, font=('arial', 15, 'bold'), text=" Clear All ", padx=2, pady=2, bg="#262526", fg="white", command=clear_all)  # Change background color
clear_button.grid(row=20, column=0, pady=10)  # Add padding

root.mainloop()
```
