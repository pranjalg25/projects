```python
import time
import random

# Declare player stats as global variables
player_health = 100
inventory = []

def introduction():
    print("Welcome to the Quest for the Enchanted Amulet!")
    time.sleep(1)
    print("You stand in the village square, eager to begin your adventure.")

def make_choice(choices):
    print("\nChoose your path:")
    for i, choice in enumerate(choices, 1):
        print(f"{i}. {choice}")

    while True:
        try:
            choice = int(input("Enter the number of your choice: "))
            if 1 <= choice <= len(choices):
                return choice
            else:
                print("Invalid choice. Try again.")
        except ValueError:
            print("Invalid input. Enter a number.")

def explore_village():
    global player_health
    print("You decide to explore the village first.")
    time.sleep(1)
    print("In the market, you come across a mysterious fortune teller.")
    time.sleep(1)

    # Random event: Receive a prophecy
    if random.choice([True, False]):
        print("The fortune teller reads your destiny and warns you of impending danger!")
        player_health -= 5

    print(f"Your current health: {player_health}")
    choices = ["Continue exploring", "Head to the Enchanted Forest"]
    return make_choice(choices)

def enter_forest():
    global player_health
    print("You enter the Enchanted Forest, surrounded by ancient trees and magical creatures.")
    time.sleep(1)
    print("As you venture deeper, you stumble upon a hidden waterfall with a mysterious glow.")
    time.sleep(1)

    # Random event: Discover a magical artifact or face a forest guardian
    if random.choice([True, False]):
        print("You discover a magical amulet hidden behind the waterfall.")
        inventory.append("Magical Amulet")
    else:
        print("A forest guardian appears, testing your courage in a friendly duel.")
        print("You gain its respect and receive a protective charm.")
        inventory.append("Protective Charm")

    print(f"Your current health: {player_health}")
    choices = ["Continue exploring", "Confront the Guardian of the Forest"]
    return make_choice(choices)

def guardian_forest_encounter():
    print("You reach a serene glade where the Guardian of the Forest awaits.")
    time.sleep(1)
    print("A mystical encounter unfolds!")

    # Random event: Impress or befriend the guardian
    if random.choice([True, False]):
        print("Your sincerity and respect impress the Guardian, who grants you a special boon.")
        print("Congratulations! You have completed your quest.")
        ending = "Victory"
    else:
        print("The Guardian tests your resolve. Although challenging, you emerge wiser.")
        print("Your quest continues with newfound wisdom.")
        ending = "Wisdom"

    return ending

def main():
    introduction()

    choices_1 = ["Explore the village", "Enter the Enchanted Forest"]
    choice = make_choice(choices_1)

    if choice == 1:
        choice = explore_village()
        if choice == 1:
            # Continue exploring the village
            print("You find a hidden library and gain knowledge about ancient artifacts.")
            inventory.append("Ancient Knowledge")
            print("Your inventory:", inventory)
            time.sleep(1)
            print("Eager to apply your newfound knowledge, you decide to head to the Enchanted Forest.")
            choice = enter_forest()
        elif choice == 2:
            # Head straight to the Enchanted Forest
            choice = enter_forest()

    elif choice == 2:
        choice = enter_forest()

    if choice == 1:
        # Continue exploring the forest
        print("You encounter a wise old owl who offers guidance.")
        time.sleep(1)
        choices_2 = ["Follow the owl's advice", "Trust your instincts"]
        choice = make_choice(choices_2)

        if choice == 1:
            print("The owl leads you to a hidden sanctuary, revealing the path to the Guardian.")
            time.sleep(1)
            choice = guardian_forest_encounter()

        elif choice == 2:
            print("You trust your instincts and navigate the forest, facing unexpected challenges.")
            time.sleep(1)
            choice = guardian_forest_encounter()

    elif choice == 2:
        # Confront the Guardian of the Forest
        choice = guardian_forest_encounter()

if __name__ == "__main__":
    main()
```
